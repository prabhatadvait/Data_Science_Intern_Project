## Level 3 Task 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('task1_level1.csv')

features = [
    "Price range", 
    "Votes", 
    "Average Cost for two", 
    "Longitude", 
    "Latitude", 
    "City", 
    "Cuisines", 
    "Has Table booking", 
    "Has Online delivery", 
    "Is delivering now"
]
x = data.loc[:, features]
y = data.iloc[:,17].values
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

x.loc[:, "Has Table booking"] = label_encoder.fit_transform(x["Has Table booking"].astype(str))
x.loc[:, "Has Online delivery"] = label_encoder.fit_transform(x["Has Online delivery"].astype(str))
x.loc[:, "Is delivering now"] = label_encoder.fit_transform(x["Is delivering now"].astype(str))

x['City'] = x['City'].apply(lambda val: val.encode('ascii', 'ignore').decode('ascii'))
x.loc[:, "City"] = x["City"].astype(str)
x.loc[:, "Cuisines"] = x["Cuisines"].astype(str)

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(
    transformers=[
        ('city_encoder', OneHotEncoder(), [3]),
        ('cuisines_encoder', OneHotEncoder(), [8])
    ],
    remainder='passthrough'
)

x = np.array(ct.fit_transform(x))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
print(x_train.shape)
print(y_train.shape)
```

```python
#Training Multiple Linear Regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# prediction the Test Result
y_predict = regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_predict.reshape(len(y_predict),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_predict)
```

## Trying SVR

```python
from sklearn.svm import SVR
regressor2 = SVR(kernel = 'rbf')
regressor2.fit(x_train, y_train)
```

```python
y_predict2 = regressor2.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_predict2.reshape(len(y_predict2),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_predict2)
```

# RAndom Forest

```python
from sklearn.ensemble import RandomForestRegressor
regressor3 = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor3.fit(x_train, y_train)
```

```python
y_predict3 = regressor3.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_predict3.reshape(len(y_predict3),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_predict3)
# this is the best model so RandomFores is best as it has accuracy of 95 %
```

# Decision Tree 

```python
from sklearn.tree import DecisionTreeRegressor
regressor4 = DecisionTreeRegressor(random_state = 0)
regressor4.fit(x_train, y_train)
```

```python
y_predict4 = regressor4.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_predict4.reshape(len(y_predict4),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_predict4)
```

## so Random Forest has highest accuracy as 95 % and then DecisionTreeRegressor has 92 %

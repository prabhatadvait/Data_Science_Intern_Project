```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('task1_level1.csv')
```

```python
# 1. Visualise the location of Restaurants
import plotly.express as px
fig = px.scatter_geo(
        data, 
        lat='Latitude', 
        lon='Longitude', 
        hover_name='City', 
        color='Aggregate rating',
        title='Restaurant Locations and Ratings',
        color_continuous_scale='Viridis'
    )
fig.show()
fig.write_html('restaurant_map.html')
```

```python
# 2. Analyze the Distribution of Restaurants across cities
top_cities = data['City'].value_counts().head(5)
print('Highest no. of Restaurant:\n',top_cities)
sns.barplot(x=top_cities.values,y=top_cities.index,color='pink')
plt.title('Top 5 Cities')
plt.xlabel('No .of Restaurant')
plt.ylabel('City')
plt.show()
```

```python
# 3. Determine Correlation b/w location and rating
print(data[['Latitude','Longitude','Aggregate rating']].corr())
print('\n')
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='Latitude',
    y='Longitude',
    hue='Aggregate rating',
    size='Aggregate rating',
    sizes=(20,200),
    palette='viridis',
    data=data
)
plt.title('Restaurant Rating Based on the Location')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()
```

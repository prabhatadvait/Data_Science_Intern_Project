```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('task1_level1.csv')

# Cuisine Popularity - Number of Votes by Cuisine
cuisine_popularity = data.groupby('Cuisines')['Votes'].sum().sort_values(ascending=False)
print(cuisine_popularity)

# Cuisine Rating Analysis - Average Rating by Cuisine
cuisine_ratings = data.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
print(cuisine_ratings)

# Highest popular cuisine is listed in the output : North Indian, Mughlai    53747
# Cuisines with highest rating are also mentioned in the output : Burger, Bar Food, Steak           4.9
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=cuisine_popularity.index, y=cuisine_popularity.values)
plt.title('Cuisine Popularity (by Number of Votes)')
plt.xlabel('Cuisine Type')
plt.ylabel('Total Number of Votes')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=cuisine_ratings.index, y=cuisine_ratings.values)
plt.title('Average Ratings by Cuisine')
plt.xlabel('Cuisine Type')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.show()
```

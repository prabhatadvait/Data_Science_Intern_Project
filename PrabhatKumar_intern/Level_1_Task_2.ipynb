```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('task1_level1.csv')
```

```python
num_col = data.select_dtypes(include=['number']).columns
print(data[num_col].describe())
print('\nmedian values :\n',data[num_col].median())
```

```python
category_column = ['Country Code','City','Cuisines']
for column in category_column:
    if column in data.columns:
        print(f'\nvalue for {column}')
        print(data[column].value_counts().head(10))
    else:
        print('column is not there')
```

```python
# Visualisation
category_column = ['Country Code','City','Cuisines']
for column in category_column:
    if column in data.columns:
        top_val = data[column].value_counts().head(10)
        sns.barplot(x=top_val.values,y=top_val.index,palette='viridis')
        plt.title(f'Top 10 {column}')
        plt.xlabel('Frequency')
        plt.ylabel(column)
        plt.show()
```

```python
top_cuisines = data['Cuisines'].value_counts().head(10)
print('Cuisines with highest no. of Restaurant',top_cuisines)
top_cuisines.plot(kind='barh',color='skyblue')
plt.title('Top 10 Cuisines')
plt.xlabel('No. of Restaurant')
plt.ylabel('Cuisines')
plt.show()
```

```python
top_cities = data['City'].value_counts().head(10)
print('Highest no. of Restaurant:\n',top_cities)
top_cities.plot(kind='barh',color='pink')
plt.title('Top 10 Cities')
plt.xlabel('No .of Restaurant')
plt.ylabel('City')
plt.show()
```

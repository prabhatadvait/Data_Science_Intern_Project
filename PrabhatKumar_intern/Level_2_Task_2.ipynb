```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('task1_level1.csv')
```

```python
# most_common = max(data.groupby('Price range').value_counts())
# print(most_common)
most_common_price_range = data['Price range'].value_counts().idxmax()
print(most_common_price_range)
```

```python
avg_rating_price = data.groupby('Price range')['Aggregate rating'].mean()
print(avg_rating_price)
```

```python
Price_range_high_avg_rating = avg_rating_price.idxmax()
high_avg_rating = avg_rating_price.max()
print(Price_range_high_avg_rating)
print(f"{high_avg_rating:.2f}")
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x=avg_rating_price.index, y = avg_rating_price.values,palette='coolwarm')
plt.bar(Price_range_high_avg_rating,high_avg_rating,color='red')
plt.title('Average Rating by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Avg_Rating')
plt.show()
```
